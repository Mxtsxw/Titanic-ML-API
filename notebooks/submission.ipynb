# Titanic: Machine Learning from Disaster

This notebook is used to make predictions for the [Titanic: Machine Learning from Disaster](https://www.kaggle.com/competitions/titanic/overview) competition on Kaggle. The predictions are saved in a CSV file that can be submitted to the competition.

---

```python
import pandas as pd 
import numpy as np
import joblib
```

```python
# Load the data
data = pd.read_csv('../data/test.csv')
```

```python
data.head(10)
```

```python
# Load the model
model = joblib.load('../artifacts/pipeline.pkl')
```

```python
model
```

```python
# Make predictions
predictions = model.predict(data)
```

```python
predictions
```

```python
# Save the predictions
res = pd.DataFrame(predictions, columns=['Survived'])
```

```python
# merge PassengerId with the predictions
res = pd.concat([data['PassengerId'], res], axis=1)
```

```python
res
```

```python
res.to_csv('../data/submission.csv', index=False)
```
